## Matriz
Una matriz A de mxn es un arreglo rectangular de mn numeros dispiestos en m renglones y n columnas. Los coeficientes de las variables x_1,x_2, x_3 en un sistema se pueden escribir como los elementos de una matriz A llamada **matriz de coeficientes** del sistema

Para poder trabajar con matrices o arrlegos, la libreria que se necesita es **Numpy**

```python
import numpy as np
```

Si tenemos el sistema de ecuaciones

2x+4y+6z=18

4x+5y+6z=24

3x+y-2z=4

La matriz de coeficientes A seria

```python
A=np.array(
    [[2,4,6],
    [4,5,6],
    [3,1,-2]]
)
```

```python
A
```

Entonces el vector de soluciones seria b

```python
b=np.array([
    [18],
    [24],
    [4]
    ])
```

```python
b
```

El sistema entonces lo podemos representar como

**Ax=b**

## Operaciones de matrices


### Suma de matrices
La suma de matrices A y B es la matriz mxn, la suma A+B se obtiene al sumar las componentes correspondientes de A y B

```python
A+matriz_identidad
```

### Teorema
Sean A, B y C tres matrices de mxn y sean alpha y beta dos escalares. Entonces:
1. A+0=A
2. 0A=0
3. A+B=B+A
4. (A+B)+C=A+(B+C)
5. alpha(A+B)=(alpha)A+(alpha)B
6. 1A=A
7. (alpha + beta)A=(alpha)A + (beta)A

### Multiplicacion de una matriz por un escalar
Si A es una matriz mxn y si alpha es un escalar, entonces la matriz mxn, (alpha)A es la matriz obtenida al multiplicar cada componente de A por alpha

```python
A
```

```python
B=3*A
B
```

### Teorema
Sean a, b y c tres vectores de dimension n y sea alpha un escalar. Entonces
1. a.0=0
2. a.b=b.a
3. a.(b+c)=a.b+a.c
4. ((alpha)a).b=alpha(a.b)

### Inversa de una matriz
Sea A y B dos matrices de nxn. Suponga que

AB=BA=I

Entonces B se llama la inversa de A y se denota por A^-1. Entonces se tiene

AA^-1=A^-1A=I

Si A tiene inversa, entonces se dice que A es invertible

```python
A=np.array([
    [2,5],
    [1,3]
])
```

```python
A_inversa=np.linalg.inv(A)
A_inversa
```

```python
A@A_inversa
```

Veamos un ejemplo del uso de una matriz inversa para resolver un sistema de ecuaciones Ax=b

2x+4y+3z=6

y-z=-4

3x+5y+7z=7

```python
A=np.array([
    [2,4,3],
    [0,1,-1],
    [3,5,7]
])
b=np.array([
    [6],
    [-4],
    [7]
])
```

```python
A_inversa=np.linalg.inv(A)
A_inversa
```

```python
x=A_inversa@b
x
```

Entonces, tenemos que
x=25, y=-8 y z=-4

Sin embargo, exite una linea para resolver el sistema sin utilizar la inversa

```python
solucion=np.linalg.solve(A,b)
solucion
```

```python
A
```

```python
A_transpuesta=A.T
A_transpuesta
```
